```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

```

```python
# Define dataset path
train_dir = "/content/drive/MyDrive/2025 MACHINE LEARNING PROJECTS/AFIT KADUNA/MODEL/DATASET/train"
test_dir = "/content/drive/MyDrive/2025 MACHINE LEARNING PROJECTS/AFIT KADUNA/MODEL/DATASET/test"

```

```python
# Define parameters
img_size = 48
batch_size = 64

```

```python
# Data Augmentation for better generalization
datagen = ImageDataGenerator(rescale=1./255,
                             rotation_range=30,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             fill_mode="nearest")

```

```python
# Load train and test datasets
train_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    color_mode="grayscale",
    class_mode="categorical"
)

test_generator = datagen.flow_from_directory(
    test_dir,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    color_mode="grayscale",
    class_mode="categorical"
)
```

```python
# Build the CNN Model
```

```python
# Build CNN Model
model = Sequential()

# Convolutional layers
model.add(Conv2D(32, (3,3), activation="relu", input_shape=(img_size, img_size, 1)))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64, (3,3), activation="relu"))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(128, (3,3), activation="relu"))
model.add(MaxPooling2D(2,2))

# Fully Connected Layers
model.add(Flatten())
model.add(Dense(128, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(7, activation="softmax"))  # 7 emotions

# Compile the model
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

# Summary
model.summary()
```

```python

# Evaluate on the test set
test_loss, test_acc = model.evaluate(test_generator)
print(f"Test Accuracy: {test_acc * 100:.2f}%")
```

```python
# Test the model on an image
# Load model

from tensorflow.keras.models import load_model

model = load_model("/content/drive/MyDrive/2025 MACHINE LEARNING PROJECTS/AFIT KADUNA/MODEL/emotion_model.h5")

# Load test image
image_path = "/content/drive/MyDrive/2025 MACHINE LEARNING PROJECTS/AFIT KADUNA/MODEL/DATASET/test/disgust/PublicTest_99162116.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (48, 48))
image = image / 255.0
image = np.expand_dims(image, axis=0)
image = np.expand_dims(image, axis=-1)

# Make prediction
predictions = model.predict(image)
emotion_labels = ["Angry", "Disgust", "Fear", "Happy", "Neutral", "Sad", "Surprise"]
emotion = emotion_labels[np.argmax(predictions)]
print(f"Predicted Emotion: {emotion}")
```
